```python
#import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']= (12,8)

# read in data

df= pd.read_csv('/Users/ryansmith/Desktop/movies.csv')


```

```python
df.head()
```

```python
# looking for missing data
for col in df.columns:
    pct = np.mean(df[col].isnull())
    print('{}-{}%'.format(col, pct))
```

```python
#getting rid of incomplete data
df.dropna(thresh=1)
```

```python
for col in df.columns:
    pct = np.mean(df[col].isnull())
    print('{}-{}%'.format(col, pct))
```

```python
df.dropna()
```

```python
df =df.dropna()
```

```python
for col in df.columns:
    pct = np.mean(df[col].isnull())
    print('{}-{}%'.format(col, pct))
```

```python
df.dtypes
```

```python
#adjusting floats to ints
df['budget']= df['budget'].astype('int64')
df['votes']= df['votes'].astype('int64')
df['gross']= df['gross'].astype('int64')
df['runtime']= df['runtime'].astype('int64')
```

```python
df.head()
```

```python
#ordering by gross revenue
df.sort_values(by=['gross'], inplace= False, ascending = False)
```

```python
#delete duplicates

df['name'].drop_duplicates().sort_values(ascending = False)
```

```python
plt.scatter(x= df['budget'], y = df['gross'])
plt.title('Budget vs Gross Earnings')
plt.ylabel('Gross Earnings')
plt.xlabel('Budget')
plt.show()
```

```python
# Plot budget vs gross using seaborn
sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws={"color" : "red"}, line_kws={"color":"blue"})
```

```python
df.corr(method='pearson')
```

```python
# high correlation between budget and gross
```

```python
correlation_matrix = df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')
plt.show()
```

```python
# look at company 
df_numerized = df
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype=='object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name]= df_numerized[col_name].cat.codes
df_numerized
```

```python
correlation_matrix = df_numerized.corr(method='pearson')
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.ylabel('Movie Features')
plt.xlabel('Movie Features')
plt.show()
```
